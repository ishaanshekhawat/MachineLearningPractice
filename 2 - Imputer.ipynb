```python
from sklearn.impute import SimpleImputer
import numpy as np
import pandas as pd
```

```python
a = np.array([[10, 200, 0.3],
              [9, np.nan, 0.1],
              [8, 240, np.nan],
              [9.5, 225, 0.45]])
pa = pd.DataFrame(a, columns=['X1', 'X2', 'X3'])
print("Means:\n", pa.mean())
```

```python
imputer = SimpleImputer().set_output(transform='pandas')
imputer.fit(pa) # computes mean or median
print(imputer.statistics_)
```

```python
imputer.transform(pa)
```
