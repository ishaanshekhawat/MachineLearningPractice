```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from tqdm import tqdm
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_absolute_error
import os
os.chdir("D:/Training/Academy/ML(Python)/Cases/Chemical Process Data")
```

```python
chem = pd.read_csv("ChemicalProcess.csv")
X, y = chem.drop('Yield', axis=1), chem['Yield']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=25)
```

```python
imputer = SimpleImputer(strategy='mean')
scaler = StandardScaler()
X_trn_imp = imputer.fit_transform(X_train)
X_tst_imp = imputer.transform(X_test)
knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(X_trn_imp, y_train)
y_pred = knn.predict(X_tst_imp)
mean_absolute_error(y_test, y_pred)
```

Using `Pipeline`

```python
imputer = SimpleImputer(strategy='mean')
pipe = Pipeline(steps=[('IMP',imputer),('KNN',knn)])
pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)
mean_absolute_error(y_test, y_pred)
```

```python
imputer = SimpleImputer(strategy='mean')
scaler = StandardScaler()
pipe = Pipeline(steps=[('IMP',imputer),('SCl',scaler),('KNN',knn)])
pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)
mean_absolute_error(y_test, y_pred)
```
