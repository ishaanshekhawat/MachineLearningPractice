```python
from transformers import pipeline
import pandas as pd
from tqdm import tqdm
```

```python
df = pd.read_csv("C:/Training/AV/Black Friday/train.csv")
df_ss = df[['User_ID','Product_ID','Purchase']]
df_ss = df_ss.astype(str)
df_ss = df_ss.head(50)
```

```python
df_ss
```

```python
tqa = pipeline(task="table-question-answering", 
               model="google/tapas-large-finetuned-wtq",
               token="ENTER_HF_TOKEN")
```

```python
question = "What is purchase for user id 1000010 and product id P00148642?"
```

```python
print(tqa(table=df_ss, query=question)['cells'][0])
```
