```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from tqdm import tqdm
from sklearn.ensemble import RandomForestRegressor, StackingRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.pipeline import Pipeline
import matplotlib.pyplot as plt
```

```python
conc = pd.read_csv("D:/Training/Academy/ML(Python)/Cases/Concrete Strength/Concrete_Data.csv")
X, y = conc.drop('Strength', axis=1), conc['Strength']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=25)
```

```python
lr = LinearRegression()
dtr = DecisionTreeRegressor(random_state=25)
rf = RandomForestRegressor(random_state=25)
ridge = Ridge()
lasso = Lasso()
el = ElasticNet()
x_gbm = XGBRegressor(random_state=25)
l_gbm = LGBMRegressor(random_state=25, verbose=-1)
c_gbm = CatBoostRegressor(random_state=25, verbose=False)
```

```python
final_est = [rf, x_gbm, l_gbm, c_gbm]
pt = [True, False]
scores = []
for f in tqdm(final_est):
    for p in pt:
        stack = StackingRegressor(estimators=[('LR',lr),('RIDGE',ridge),('LASSO',lasso),
                                              ('EL', el),('TREE', dtr)], 
                                  final_estimator=f, passthrough=p)
        stack.fit(X_train, y_train)
        y_pred = stack.predict(X_test)
        scores.append([f,p,mean_absolute_error(y_test, y_pred)])
df_scores = pd.DataFrame(scores, columns=['final_est', 'passthrough', 'score'])
df_scores.sort_values('score')
```

```python
final_est = [rf, x_gbm, l_gbm, c_gbm]
pt = [True, False]
scores = []
for f in tqdm(final_est):
    for p in pt:
        stack = StackingRegressor(estimators=[('LR',lr),('RIDGE',ridge),('LASSO',lasso),
                                              ('EL', el),('TREE', dtr)], 
                                  final_estimator=f, passthrough=p)
        stack.fit(X_train, y_train)
        y_pred = stack.predict(X_test)
        scores.append([f,p,r2_score(y_test, y_pred)])
df_scores = pd.DataFrame(scores, columns=['final_est', 'passthrough', 'score'])
df_scores.sort_values('score', ascending=False)
```
