```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import pandas as pd
```

```python
a = np.array([[10, 200, 0.3],
              [9, 230, 0.1],
              [8, 240, 0.24],
              [9.5, 225, 0.45]])
pa = pd.DataFrame(a, columns=['X1', 'X2', 'X3'])
print("Means:\n", pa.mean())
print("Std Deviations:\n",pa.std(ddof=0))
```

```python
scaler = StandardScaler().set_output(transform='pandas')
scaler.fit(pa)
print(scaler.mean_)
print(scaler.scale_)
```

```python
scaler.transform( pa )
```

```python
scaler.fit_transform(pa)
```

#### Min Max

```python
scaler = MinMaxScaler().set_output(transform='pandas')
scaler.fit(pa)
print( scaler.data_min_ )
print( scaler.data_max_ )
```

```python
scaler.transform(pa)
```
