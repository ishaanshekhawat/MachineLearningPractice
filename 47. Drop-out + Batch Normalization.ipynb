```python
import pandas as pd
import tensorflow as tf
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import log_loss, accuracy_score
import os
os.chdir("D:/Training/Academy/ML(Python)/Datasets")
```

```python
wine = pd.read_csv("wine.csv")
X, y = wine.drop('Class', axis=1), wine['Class']-1
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=25,
                                                   stratify=y)
scl_x = MinMaxScaler()
X_trn_scl, X_tst_scl = scl_x.fit_transform(X_train), scl_x.transform(X_test)
```

```python
intializer = tf.keras.initializers.GlorotUniform(seed = 25)
tf.random.set_seed(25)
model = tf.keras.Sequential([tf.keras.Input(shape=(X_trn_scl.shape[1],)),
                             tf.keras.layers.Dense(50, activation='relu',
                                                   kernel_initializer=intializer),
                             tf.keras.layers.Dense(20, activation='relu',
                                                   kernel_initializer=intializer),
                             tf.keras.layers.Dense(10, activation='relu',
                                                   kernel_initializer=intializer),
                             tf.keras.layers.Dense(3, activation='softmax',
                                                   kernel_initializer=intializer)])
model.compile(optimizer=tf.keras.optimizers.Adamax(learning_rate=0.001),
               loss=tf.keras.losses.SparseCategoricalCrossentropy(), 
               metrics=['acc'])
history = model.fit(X_trn_scl, y_train,validation_data=(X_tst_scl,y_test), epochs=100)
```

```python
y_pred_prob = model.predict(X_tst_scl)
print("Log Loss =",log_loss(y_test, y_pred_prob))
y_pred = y_pred_prob.argmax(axis=1)
print("Accuracy =",accuracy_score(y_test, y_pred))
```

**Drop Out**

```python
intializer = tf.keras.initializers.GlorotUniform(seed = 25)
tf.random.set_seed(25)
model = tf.keras.Sequential([tf.keras.Input(shape=(X_trn_scl.shape[1],)),
                             tf.keras.layers.Dropout(0.1),
                             tf.keras.layers.Dense(50, activation='relu',
                                                   kernel_initializer=intializer),
                             tf.keras.layers.Dropout(0.2),
                             tf.keras.layers.Dense(20, activation='relu',
                                                   kernel_initializer=intializer),
                             tf.keras.layers.Dropout(0.1),
                             tf.keras.layers.Dense(10, activation='relu',
                                                   kernel_initializer=intializer),
                             tf.keras.layers.Dense(3, activation='softmax',
                                                   kernel_initializer=intializer)])
model.compile(optimizer=tf.keras.optimizers.Adamax(learning_rate=0.001),
               loss=tf.keras.losses.SparseCategoricalCrossentropy(), 
               metrics=['acc'])
history = model.fit(X_trn_scl, y_train,validation_data=(X_tst_scl,y_test), epochs=100)
```

```python
y_pred_prob = model.predict(X_tst_scl)
print("Log Loss =",log_loss(y_test, y_pred_prob))
y_pred = y_pred_prob.argmax(axis=1)
print("Accuracy =",accuracy_score(y_test, y_pred))
```

**Batch Normalization**

```python
intializer = tf.keras.initializers.GlorotUniform(seed = 25)
tf.random.set_seed(25)
model = tf.keras.Sequential([tf.keras.Input(shape=(X_trn_scl.shape[1],)),
                             tf.keras.layers.BatchNormalization(),
                             tf.keras.layers.Dense(50, activation='relu',
                                                   kernel_initializer=intializer),
                             tf.keras.layers.BatchNormalization(),
                             tf.keras.layers.Dense(20, activation='relu',
                                                   kernel_initializer=intializer),
                             tf.keras.layers.BatchNormalization(),
                             tf.keras.layers.Dense(10, activation='relu',
                                                   kernel_initializer=intializer),
                             tf.keras.layers.BatchNormalization(),
                             tf.keras.layers.Dense(3, activation='softmax',
                                                   kernel_initializer=intializer)])
model.compile(optimizer=tf.keras.optimizers.Adamax(learning_rate=0.001),
               loss=tf.keras.losses.SparseCategoricalCrossentropy(), 
               metrics=['acc'])
history = model.fit(X_trn_scl, y_train,validation_data=(X_tst_scl,y_test), epochs=100)
```

```python
y_pred_prob = model.predict(X_tst_scl)
print("Log Loss =",log_loss(y_test, y_pred_prob))
y_pred = y_pred_prob.argmax(axis=1)
print("Accuracy =",accuracy_score(y_test, y_pred))
```
